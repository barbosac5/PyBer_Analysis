```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jly", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.bar(x_axis, y_axis, color = "green",  label = 'Boston')

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot (Using GCA method to invert axis)
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot and add a label for the legend
plt.barh(x_axis, y_axis, color = "magenta",  label = 'Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axis
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Set the y limit between 0 and 45
plt.xlim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month (USING OBJECT ORIENTED APPROACH)
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jly", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
ax.bar(x_axis, y_axis, color = "black", label = "New York")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (Reversed axis)
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Skill Drill 
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fares by Month")
```
