```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month (USING OBJECT ORIENTED APPROACH)
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jly", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a plot
plt.scatter(x_axis, y_axis)
```

```python
# Skill Drill
# Create a plot
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.legend()
plt.xlabel("Month")
plt.ylabel("Fare ($)")
plt.title("PyBer Fares by Month")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
# Changing the size of the dots 
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Adjusting the size of the bubbles
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Form the plot
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# List Comprehension 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter plot with Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble Chart with Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Skill Drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis])
ax.scatter(y_axis, x_axis, edgecolor = "black", color = "skyblue", linewidth = 2, alpha = 0.25, label = "Boston",)
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fares by Month")
ax.set_xlim(0, 50)
fig.legend()
ax.invert_yaxis()
```
